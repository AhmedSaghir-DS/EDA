### Feature Engineering

```python
#for better readability
df.rename(columns = {"average_cost_for_two":"cost2plates"}, inplace = True)
df.head(1)
```

```python
df['establishment'].unique()
```

```python
# Found empty brackets. replacing it with Not Specified
df['establishment'] = df['establishment'].str.replace('[]', '[Not Specified]')
df['establishment'].unique()
```

```python
# Strip square brackets and quotes from the 'establishment' column
#df['establishment'] = df['establishment'].str.strip("[]").replace("'", "")

# Cleaning the 'establishment' column by removing square brackets and single quotes
df['establishment'] = df['establishment'].str.replace(r"[\[\]']", "", regex=True)
df['establishment'].unique()
```

Looks like there are rating in other languages aswell. going to add them in others for better readability

```python
df['rating_text'].value_counts()
```

```python
rating_types = df['rating_text'].value_counts(ascending  = False)
rating_types

rating_types_lessthan500 = rating_types[rating_types<500]
rating_types_lessthan500
```

```python
def handle_rating_text(value):
    if(value in rating_types_lessthan500):
        return 'others'
    else:
        return value
        
df['rating_text'] = df['rating_text'].apply(handle_rating_text)
df['rating_text'].value_counts()
```

## 5. Regional Analysis

### Restraunts Trends and Customer Preferences

```python
total_restaurants = df['city'].value_counts()
total_restaurants
```

```python
# Count the number of restaurants per city and select top 10 cities with the most listings
top_cities = df['city'].value_counts().head(10).index

# Filter data for these top cities
top_cities_data = df[df['city'].isin(top_cities)]
top_cities
```

```python
# Plotting establishment types across top cities
plt.figure(figsize=(10, 6))
sns.countplot(y='establishment', data=top_cities_data, order=top_cities_data['establishment'].value_counts().index[:10])
plt.title('Top 10 Establishment Types Across Major Cities')
plt.xlabel('Count')
plt.ylabel('Establishment Type')
plt.show()
```

```python
# Average cost for two across top cities
avg_cost_by_city = top_cities_data.groupby('city')['cost2plates'].mean().sort_values()

plt.figure(figsize=(10, 6))
avg_cost_by_city.plot(kind='bar', color='skyblue')
plt.title('Average Cost for Two Across Top Cities')
plt.xlabel('City')
plt.ylabel('Average Cost')
plt.xticks(rotation=45)
plt.show()
```

```python
# Price range across top cities
price_range_by_city = top_cities_data.groupby('city')['price_range'].mean().sort_values()

plt.figure(figsize=(10, 6))
price_range_by_city.plot(kind='bar', color='green')
plt.title('Average Price Range Across Top Cities')
plt.xlabel('City')
plt.ylabel('Average Price Range')
plt.xticks(rotation=45)
plt.show()
```

- Casual Dining and Quick Bites are the most common types of establishments across these major cities. Cafes, Dessert Parlours, and Bakeries also appear frequently.
- Mumbai has the highest average cost for two people dining, followed by New Delhi and Bangalore, indicating these cities might cater to a more upscale market.
<br> On the more affordable side, we have Ahmedabad and Lucknow, which have lower average dining costs.
- Goa stands out with a higher average price range, possibly due to its status as a tourist destination.
<br> Mumbai and New Delhi also show higher price ranges, reflecting their metro status and potentially higher living costs.
<br> Cities like Chennai, Kolkata, and Jaipur maintain a more moderate pricing profile.

### Dining Preferences

```python
# As cuisines are in a list-like structure, separating it
df['cuisines'] = df['cuisines'].str.split(', ')
cuisines_list = df.explode('cuisines')
cuisine_types = cuisines_list.groupby('city')['cuisines'].value_counts().unstack(fill_value=0)
```

```python
# Visualizing top cuisines in the top 5 cities
top_cuisines = cuisine_types.loc[total_restaurants.head(5).index]
top_5_cuisines = top_cuisines.apply(lambda x: x.nlargest(5), axis=1)
top_5_cuisines.plot(kind='bar', figsize=(14, 8), stacked=True)
plt.title('Top 5 Popular Cuisines in Each City')
plt.xlabel('City')
plt.ylabel('Number of Restaurants Offering the Cuisine')
plt.legend(title='Cuisines', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

- Here are the top three cuisines for each of the major cities

-  Chennai: North Indian, Chinese, South Indian
-  Mumbai: North Indian, Chinese, Fast Food
-  Bangalore: North Indian, Chinese, Continental
-  Pune: North Indian, Chinese, Fast Food
-  New Delhi: North Indian, Chinese, Fast Food

```python
# Aggregate rating across top cities
rating_by_city = top_cities_data.groupby('city')['aggregate_rating'].mean().sort_values()

plt.figure(figsize=(10, 6))
rating_by_city.plot(kind='bar', color='coral')
plt.title('Rating Across Top Cities')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

- Kolkata, New Delhi and Mumbai exhibit the highest average customer ratings, suggesting a high level of customer satisfaction or possibly better quality restaurants.rants.
<br> Goa has a notably lower average rating, which could reflect a variety of factors, including tourist- services variations

## 6. Customer Preference Analysis

### Top Choice Cuisines

```python
# Group by 'city' and 'cuisines' and count occurrences
cuisine_popularity = cuisines_list.groupby(['city', 'cuisines']).size().reset_index(name='count')

# Sort and find the top 5 cuisines in each city
top_cuisines_by_city_grouped = cuisine_popularity.sort_values(['city', 'count'], ascending=[True, False]).groupby('city').head(5)
```

- Varied Preferences: While North Indian cuisine remains popular across these cities, we can see a diverse range of other cuisines featuring prominently. For instance, Fast Food and Chinese are common across all cities.
- Regional Influences: Cities like Chennai and Kolkata show a strong preference for local cuisines such as South Indian in Chennai and Bengali in Kolkata, reflecting regional culinary traditions.
- International Flavors: Cities like Mumbai and Pune, known for their cosmopolitan populace, display a wider variety of cuisines including Italian and Continental, indicating a palate for international flavors.
- Unique Trends: Each city has its own unique set of popular cuisines which could be influenced by local tastes, demographics, and cultural factors. For example, Street Food is notably popular in Ahmedabad.
